```python
columns_to_remove=['date','yr_renovated','street','city','statezip','country']
df = df.drop(columns=columns_to_remove)     #Unused column to delete for analyze
```

```python
df.head()
```

```python
import scipy.stats as stats    #For Z-Index
```

```python
#For Removing Outliers

z_scores=stats.zscore(df)
threshold=3
print("Size Befor removing Outliers:",df.shape)   #df.shape is print for count rows
outliers_df=df[(z_scores>threshold).any(axis=1)]  #axis=1 column access
df=df[(z_scores <= threshold).all(axis=1)]
print("Size after removing outliers:", df.shape)   
```

```python
from sklearn.preprocessing import StandardScaler  #Convert all unit scales
```

```python
scaler = StandardScaler()

scaler.fit(df)             #Fit the df to standardscaler

df_scaled = pd.DataFrame(scaler.transform(df),columns=df.columns)    #Tranform & Tell tranform the all columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x= df.drop('price',axis=1)
y= df['price']

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(f"'xtrain {x_train.shape}")
print(f"'xtest {x_test.shape}")
```

```python
models = [
    ('Random forest', RandomForestRegressor()),
    ('Linear Regression', LinearRegression()),
    ('Decision Tree', DecisionTreeRegressor()),
    ('KNN', KNeighborsRegressor())
]
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
for name,model in models:
    print(name)
    print()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print("Mean squaed error:", mean_absolute_error(y_test, y_pred))
    print('/n')
    print("Mean Absolute Error",mean_absolute_error(y_test,y_pred))
    print('/n')
    print("R-squared (R2)",r2_score(y_test,y_pred))
    print('/n')
```
