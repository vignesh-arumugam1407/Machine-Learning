```python
text_columns=['packed_cell_volume','white_blood_cell_count','red_blood_cell_count']

for i in text_columns:                      #Iterate
    print(f"{i} : {df_data[i].dtype}")      #Show the column datatype
```

```python
#Change the Numeric Column

'''error= 'coerce' (If a value cannot be converted (like a string "abc"), it replaces it with NaN (missing value).)
pd.to_numeric (This tries to convert the values in that column to numbers (integers or floats).)
'''

def convert_text_to_numeric(df_data,column):
    df_data[column] = pd.to_numeric(df_data[column], errors='coerce')


for column in text_columns:
    convert_text_to_numeric(df_data,column)
    print(f"{column} : {df_data[column].dtype}")
```

```python
missing = df_data.isnull().sum()     #For print total number of null values

#Print above 0 with Descending Order by first 20 Values
missing[missing > 0].sort_values(ascending=False).head(20) 
```

```python
# For Numerical data

def mean_value_imputation(df_data,column):
    mean_value=df_data[column].mean()      #Found Mean Value
    df_data[column].fillna(value=mean_value, inplace=True)  #inplace=True means it updates the original DataFrame directly

# For categorical or repeated data
 
def mode_value_imputation(df_data,column):
    mode=df_data[column].mode()[0]      #[0] gets the first (most common) one.
    df_data[column]=df_data[column].fillna(mode)

```

```python
df_data.columns
```

```python
# We do arrange numerical & categorical data into array for store.

#List comprehension

'''
for col in df_data.columns
This starts a loop that goes through each column name one by one.

'object' â†’ usually text/string

'int64' or 'float64' â†’ numbers

if df_data[col].dtype != 'object'
This checks: "Is the column not of type 'object'?"

If yes (i.e., it's numeric), keep it.
'''

num_cols= [col for col in df_data.columns if df_data[col].dtype != 'object']
```

```python
for col_name in num_cols:
    mean_value_imputation(df_data,col_name)
```

```python
cat_cols= [col for col in df_data.columns if df_data[col].dtype == 'object']
for col_name in cat_cols:
    mode_value_imputation(df_data,col_name)
```

```python
df_data['diabetes_mellitus'] = df_data['diabetes_mellitus'].replace(to_replace={' yes':'yes' , '\tno':'no' , '\tyes':'yes'})
df_data['coronary_artert_disease'] = df_data['coronary_artert_disease'].replace(to_replace={'\tno':'no'})
df_data['class'] = df_data['class'].replace(to_replace={'ckd\t':'ckd' ,'notckd':'not ckd'})
```

```python
#For Check
print(f"diabetes_mellitus :- {df_data['diabetes_mellitus'].unique()}")
print(f"coronary_artert_disease :- {df_data['coronary_artert_disease'].unique()}")
print(f"class :- {df_data['class'].unique()}")
print(f"hypertension :- {df_data['hypertension'].unique()}")
print(f"red_blood_cells :- {df_data['red_blood_cells'].unique()}")
print(f"pus_cells :- {df_data['pus_cells'].unique()}")
print(f"pus_cell_clumps :- {df_data['pus_cell_clumps'].unique()}")
print(f"bacteria :- {df_data['bacteria'].unique()}")
print(f"appetite :- {df_data['appetite'].unique()}")
print(f"peda_edema :- {df_data['peda_edema'].unique()}")
print(f"anemia :- {df_data['anemia'].unique()}")
```

```python
#Change to Categorical value to Numerical value

df_data['diabetes_mellitus']=df_data['diabetes_mellitus'].map({'yes':1, 'no':0})
df_data['coronary_artert_disease']=df_data['coronary_artert_disease'].map({'yes':1, 'no':0})
df_data['class']=df_data['class'].map({'ckd':1, 'not ckd':0})
df_data['hypertension']=df_data['hypertension'].map({'yes':1, 'no':0})
df_data['red_blood_cells']=df_data['red_blood_cells'].map({'normal':1, 'abnormal':0})
df_data['pus_cells']=df_data['pus_cells'].map({'normal':1, 'abnormal':0})
df_data['pus_cell_clumps']=df_data['pus_cell_clumps'].map({'present':1, 'notpresent':0})
df_data['bacteria']=df_data['bacteria'].map({'present':1, 'notpresent':0})
df_data['appetite']=df_data['appetite'].map({'good':1, 'poor':0})
df_data['peda_edema']=df_data['peda_edema'].map({'yes':1, 'no':0})
df_data['anemia']=df_data['anemia'].map({'yes':1, 'no':0})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_data.drop('class',axis=1)  # All columns without class column
y = df_data['class']

x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(f"'X_train' Shape : { x_train.shape}")
print(f"'X_test' Shape : { x_test.shape}")
```

```python
from sklearn.tree import DecisionTreeClassifier

dct = DecisionTreeClassifier()
dct.fit(x_train,y_train)
```

```python
y_pred_dct = dct.predict(x_test)

y_pred_dct
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
models =[]
models.append(('naive bayes', GaussianNB()))
models.append(('KNN', KNeighborsClassifier(n_neighbors=8)))
models.append(('Random forest Classifier', RandomForestClassifier()))
models.append(('Decisiontree Classifier', DecisionTreeClassifier()))
models.append(('SVM', SVC(kernel='linear')))
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
for name,model in models:
    print(name, model)
    print()

    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)

    print('Confusion Matrix :' , confusion_matrix(y_test,y_pred))
    print('/n')
    print('Accuracy :' , accuracy_score(y_test,y_pred))
    print('/n')
    print('Precision :' , precision_score(y_test,y_pred))
    print('/n')
    print('Recall :' , recall_score(y_test,y_pred))
    print('/n')
    print('f1 Score :' , f1_score(y_test,y_pred))
    print('/n')
```
