Series is an One Dimensional Array, then it is a homogenius array, it stored only one datatype. Like only integer or only strings. We can change the index.

```python
data = [10,20,30]
series = pd.Series(data)
series
```

```python
data = [10,20,30]
custom_index = ['A','B','C']       # Change the index
series = pd.Series(data, index=custom_index)
series
```

```python
series['B']         #-- Output is 20  ( we can access throgh the custom index)
series[1]           #-- Output is 20  ( we can access throgh the Default index)
```

```python
#For Adding Columns 

df['Rating'] = [1,2,4,5,6,7,4,1,2,5,8,4,1,2,3,4,5,9,7,4,1,2,1,5,8,9,6,5,1,2,3]
df
```

```python
group = df.groupby('Rating')['Calories'].mean()  
#We Create Rating wise group and we find the mean of every calories in group
group
```

```python
#Exclude non-numeric columns
#If you just want correlations among numeric columns:

df_numeric = df.select_dtypes(include='number')
''' 
-  Filters the original DataFrame df to include only the columns with numeric data types (like int, float, int64, etc.).
-  This creates a new DataFrame called df_numeric, removing any non-numeric columns such as strings, dates, or categorical labels.
-  This step is important because correlation calculations only work with numerical data.

'''
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
#Convert the date column to datetime and then to numeric

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
''' 
-  Purpose: Converts the Date column (which contains string dates like '2020/12/01') into proper datetime objects.
-  errors='coerce': If a date string is invalid or unrecognized, it turns that entry into NaT (Not a Time), avoiding errors.
-  Now, your Date column is ready for time-based operations and analysis

'''
df['Date_numeric'] = df['Date'].astype('int64')  # Converts to nanoseconds since epoch
''' 
-  Converts each datetime entry into an integer using astype('int64').
-  These integers represent the number of nanoseconds since the Unix Epoch (1970-01-01).
-  This turns your dates into usable numeric values, allowing them to participate in mathematical and statistical functions like correlation.

'''
correlation_matrix1 = df.corr()         #Fine the correlation
correlation_matrix1
```
