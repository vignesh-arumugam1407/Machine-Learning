```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4,5,6])

y= 2*x
y
```

```python
plt.plot(x,y, color='orange', linestyle=':',linewidth=3)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
z=3*x
z
```

```python
plt.plot(x,y, color='orange', linestyle=':',linewidth=2)
plt.plot(x,z, color='red', linestyle='-.',linewidth=2)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)     #1 Row 2 Columns and this plot index is 1.
plt.plot(x,y, color='orange', linestyle=':',linewidth=2)

plt.subplot(1,2,2)     #1 Row 2 Columns and this plot index is 2.
plt.plot(x,z, color='blue', linestyle='-.',linewidth=2)

plt.show()
```

```python
student = {
    'Bob': 68,
    'Lil Wayne' : 60,
    'Wiz' : 63,
    'Eminem' : 90 
}
```

```python
name = list(student.keys())
value = list(student.values())
```

```python
plt.bar(name,value)
plt.title('Bar Plot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
plt.barh(name,value, color='g')
plt.title('Horizontal Bar Plot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
a=[10,20,30,40,50,60,70,80,90]
b=[5,7,8,4,6,9,1,3,2]
c=[8,5,2,1,7,4,6,9,3]

plt.scatter(a,b, marker='*', color='g', s=100)
plt.scatter(a,c, marker='.', color='r', s=150)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(a,b, marker='*', color='g', s=100)

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(1,2,2)
plt.scatter(a,c, marker='.', color='r', s=150)

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
